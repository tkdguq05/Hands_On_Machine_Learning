```python
cwd = os.getcwd()
path = cwd + "/img/"
flist = os.listdir(path)
print("{0} FILE ARE IN {1}".format(len(flist), path))
```

```python
# LIST UP FILES INSIDE THE FOLDER
for i, f in enumerate(flist):
    print ("{0} TH FILE IS {1}".format(i, f))
```

```python
import imageio
```

```python
# LOAD IMAGES WITH APPROPRIATE EXTENSIONS
valid_exts = [".jpg", ".gif", ".png", ".tga", ".jpeg"]
imgs = []
names = []
for f in flist:
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_exts:
        continue
    # LOAD AND APPEND TO IMGS LIST
    fullpath = os.path.join(path, f)
    imgs.append(imageio.imread(fullpath))
    # APPEND NAMES
    names.append(os.path.splitext(f)[0])
```

# PNG 이미지가 보이지 않을 때

png파일에 RGBA형태로 저장되어 있을 경우 imageio에서는 [255,255,255,255]형태로 되어있어 읽어내지 못한다.  
이럴때는 아래의 코드를 이용하여 파일의 형식을 변경하고 새로운 파일을 만들어 읽히도록 한다.

```python
dicepath ='/Users/sanghyub/space/Hands_On_Machine_Learning/img/dices.png'
```

```python
from PIL import Image

png = Image.open(dicepath).convert('RGBA')
background = Image.new('RGBA', png.size, (255,255,255))

alpha_composite = Image.alpha_composite(background, png)
alpha_composite.save('foo.png', 'png', quality=80)
```
